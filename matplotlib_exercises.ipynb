1. Plot $y = x^{2} - x + 2$.

- Generate the x and y points.
- Annotate the origin.

```python
import matplotlib.pyplot as plt
```

```python
x = list(range(-20, 20))

y = [(n ** 2 - n + 2) for n in x]

plt.plot(x, y)
plt.title('$y = x^{2} - x + 2$')
plt.annotate('Origin', xy = (0, 0), xytext = (0, 100), arrowprops = {'facecolor': 'purple'})

plt.show()
```

2. a) Plot $y = \sqrt{x}$

```python
from math import sqrt

x = list(range(0, 20))

y = [sqrt(n) for n in x]

plt.plot(x, y)
plt.title('$y = \sqrt{x}$')
plt.xlabel('$x$')
plt.ylabel('$y$')

plt.show()
```

b) Plot $y = x^{3}$

```python
x = list(range(-20, 20))

y = [n ** 3 for n in x]

plt.plot(x, y)
plt.title('$y = x^{3}$')
plt.xlabel('$x$')
plt.ylabel('$y$')

plt.show()
```

c) Plot $y = 2^{x}$

```python
x = list(range(-20, 20))

y = [2 ** n for n in x]

plt.plot(x, y)
plt.title('$y = 2^{x}$')
plt.xlabel('$x$')
plt.ylabel('$y$')

plt.show()
```

d) y = 1/(x + 1)

```python
x = list(range(0, 20))

y = [1 / (n + 1) for n in x]

plt.plot(x, y)
plt.title('$y = {1} / {(n + 1)}$')
plt.xlabel('$x$')
plt.ylabel('$y$')

plt.show()
```

3. Combine the figures you created in the last step into one large figure with 4 subplots 

```python
plt.figure(figsize = (16, 8))

# plot $\sqrt{x}$
plt.subplot(2, 2, 1)

x = list(range(0, 50))
y = [sqrt(n) for n in x]

plt.plot(x, y)
plt.title('$\sqrt{x}$')
plt.ylabel('$y$')


# plot $x^{3}$
plt.subplot(2, 2, 2) 

x = list(range(-10, 10))
y = [n ** 3 for n in x]

plt.plot(x, y)
plt.title('$x^{3}$')
plt.xlabel('$x$')


# plot $2^{x}$
plt.subplot(2, 2, 3) 

x = list(range(-10, 10))
y = [2 ** n for n in x]

plt.plot(x, y)
plt.title('$2^{x}$')
plt.xlabel('$x$')
plt.ylabel('$y$')


# plot $y = {1} / {(n + 1)}$
plt.subplot(2, 2, 4)

x = list(range(0, 10))

y = [1 / (n + 1) for n in x]

plt.plot(x, y)
plt.title('$y = {1} / {(n + 1)}$')
plt.xlabel('$x$')
plt.ylabel('$y$')


plt.show()
```

4. Combine the figures you created in the last step into one figure where each of the 4 equations has a different color for the points. Be sure to include a legend and an appropriate title for the figure.

```python
x = list(range(0, 30))

y1 = [sqrt(n) for n in x]
y2 = [n**3 for n in x]
y3 = [2**n for n in x]
y4 = [1/(n + 1) for n in x]

plt.plot(x, y1, c='red', label='$\sqrt{x}$', alpha=.5)
plt.plot(x, y2, c='green', label='$x^{3}$', alpha=.5)
plt.plot(x, y3, c='blue', label='$2^{x}$', alpha=.5)
plt.plot(x, y4, c='black', label='${1}/{(n + 1)}$', alpha=.5)
plt.ylim(0, 20)
plt.xlim(0, 20)

plt.ylabel('$y$')
plt.xlabel('$x$')
plt.title('Combined Equations')
plt.legend()

plt.show()
```
