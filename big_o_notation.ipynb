```python
import matplotlib.pyplot as plt
import math
```

```python
plt.figure(figsize = (12, 6))

x = range(1, 20)

y1 = [1 for n in x]
y2 = [math.log(n) for n in x]
y3 = x
y4 = [n * math.log(n) for n in x]
y5 = [n**2 for n in x]
y6 = [2**n for n in x]
y7 = [math.factorial(n) for n in x]
y8 = [n**n for n in x]

plt.plot(x, y1, label = '$\mathcal{O}(1)$', color = 'red')
plt.plot(x, y2, label = '$\mathcal{O}(logn)$', color = 'black')
plt.plot(x, y3, label = '$\mathcal{O}(n)$', color = 'green')
plt.plot(x, y4, label = '$\mathcal{O}(nlogn)$', color = 'blue')
plt.plot(x, y5, label = '$\mathcal{O}(n^{2})$', color = 'purple')
plt.plot(x, y6, label = '$\mathcal{O}(2^{n})$', color = 'salmon')
plt.plot(x, y7, label = '$\mathcal{O}(n!)$', color = 'lime')
plt.plot(x, y8, label = '$\mathcal{O}(n^{n})$', color = 'turquoise')

plt.title('Big O Notation', fontsize = 16)
plt.ylabel('Operations', fontsize = 14)
plt.xlabel('Elements', fontsize = 14)

plt.xlim(1, 11)
plt.ylim(0, 50)
plt.legend()

plt.show()
```
